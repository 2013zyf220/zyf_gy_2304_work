```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon
```

```python
#up2024_0422_16:25
subtitle_1 = ['Normal summer day', 'Extremely hot day']
day_1 = [5,4]
loc_1 = 'E:/zyf_gn/zyf_gn_2301_data/ppa_2301_k2/V2/DATA_ANA_1/'

data_all = dict()
data_all[0] = pd.read_csv(loc_1 + 'ppa_2301_ana_s5.csv')
data_all[1] = pd.read_csv(loc_1 + 'ppa_2301_ana_s4.csv')

data_all[0]
```

```python
#up2024_0422_21:05
ser_1 = np.arange(30) + 0
print('ser_1: %s'%ser_1)

ser_2 = np.arange(152) + 33
print('ser_2: %s'%ser_2)

ser_all = np.append(ser_1, ser_2)
print('ser_all: %s'%ser_all)

ser_2cj = np.arange(64) + 33
print('ser_2cj : %s'%ser_2cj)

ser_jl = np.arange(88) + 97
print('ser_jl : %s'%ser_jl)

ser_cj = np.append(ser_1, ser_2cj)
print('ser_cj: %s'%ser_cj)
```

```python
#up2024_0423_08:05
ser_left = np.arange(94) + 48
print('ser_left: %s'%ser_left)

ser_right_1 = np.arange(15) + 33
print('ser_right_1: %s'%ser_right_1)

ser_right_2 = np.arange(43) + 142
print('ser_right_2: %s'%ser_right_2)

ser_right_t1 = np.append(ser_1, ser_right_1)
ser_right = np.append(ser_right_t1, ser_right_2)
print('ser_right: %s'%ser_right)
```

```python
#up2024_0422_16:26
col_1 = ['MIN','MAX','MEAN','STD','MIN_CJ','MAX_CJ','MEAN_CJ','STD_CJ','MIN_JL','MAX_JL','MEAN_JL','STD_JL',\
         'MIN_ALL','MAX_ALL','MEAN_ALL','STD_ALL','MIN_LEFT','MAX_LEFT','MEAN_LEFT','STD_LEFT','MIN_RIGHT','MAX_RIGHT','MEAN_RIGHT','STD_RIGHT']
index_y = ['XY_rci','XY_rcd','XY_crci']

data_v = dict()
data_v_cj = dict()
data_v_jl = dict()
data_v_all = dict()
data_v_left = dict()
data_v_right = dict()

data_v_sta_df = dict()
data_v_sta = dict()

for ii in range(len(index_y)):
    data_v[ii] = dict()
    data_v_cj[ii] = dict()
    data_v_jl[ii] = dict()
    data_v_all[ii] = dict()
    data_v_left[ii] = dict()
    data_v_right[ii] = dict()
    c_index = index_y[ii]
    data_v_sta[ii] = np.zeros((2, 24))
    for jj in range(2):
        data_v[ii][jj] = data_all[jj][c_index].values[ser_all]
        data_v_cj[ii][jj] = data_all[jj][c_index].values[ser_cj]
        data_v_jl[ii][jj] = data_all[jj][c_index].values[ser_jl]
        data_v_all[ii][jj] = data_all[jj][c_index].values
        data_v_left[ii][jj] = data_all[jj][c_index].values[ser_left]
        data_v_right[ii][jj] = data_all[jj][c_index].values[ser_right]
        data_v_sta[ii][jj,0] = np.min(data_v[ii][jj])
        data_v_sta[ii][jj,1] = np.max(data_v[ii][jj])
        data_v_sta[ii][jj,2] = round(np.mean(data_v[ii][jj]),3)
        data_v_sta[ii][jj,3] = round(np.std(data_v[ii][jj]),3)
        data_v_sta[ii][jj,4] = np.min(data_v_cj[ii][jj])
        data_v_sta[ii][jj,5] = np.max(data_v_cj[ii][jj])
        data_v_sta[ii][jj,6] = round(np.mean(data_v_cj[ii][jj]),3)
        data_v_sta[ii][jj,7] = round(np.std(data_v_cj[ii][jj]),3)
        data_v_sta[ii][jj,8] = np.min(data_v_jl[ii][jj])
        data_v_sta[ii][jj,9] = np.max(data_v_jl[ii][jj])
        data_v_sta[ii][jj,10] = round(np.mean(data_v_jl[ii][jj]),3)
        data_v_sta[ii][jj,11] = round(np.std(data_v_jl[ii][jj]),3)
        data_v_sta[ii][jj,12] = np.min(data_v_all[ii][jj])
        data_v_sta[ii][jj,13] = np.max(data_v_all[ii][jj])
        data_v_sta[ii][jj,14] = round(np.mean(data_v_all[ii][jj]),3)
        data_v_sta[ii][jj,15] = round(np.std(data_v_all[ii][jj]),3)
        data_v_sta[ii][jj,16] = np.min(data_v_left[ii][jj])
        data_v_sta[ii][jj,17] = np.max(data_v_left[ii][jj])
        data_v_sta[ii][jj,18] = round(np.mean(data_v_left[ii][jj]),3)
        data_v_sta[ii][jj,19] = round(np.std(data_v_left[ii][jj]),3)
        data_v_sta[ii][jj,20] = np.min(data_v_right[ii][jj])
        data_v_sta[ii][jj,21] = np.max(data_v_right[ii][jj])
        data_v_sta[ii][jj,22] = round(np.mean(data_v_right[ii][jj]),3)
        data_v_sta[ii][jj,23] = round(np.std(data_v_right[ii][jj]),3)
    data_v_sta_df[ii] = pd.DataFrame(data_v_sta[ii],  index = subtitle_1, columns = col_1)
#print(data_v[0][1])
```
